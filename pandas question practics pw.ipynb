```python
#Question 3
import pandas as pd

def sum_first_three(df):
    if len(df) < 3:
        print(df['Values'].sum())
    else:
        sum_of_first_three = df['Values'].iloc[:3].sum()
        print(sum_of_first_three)

df = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})
sum_first_three(df)
```

```python
#Qustion 8
import pandas as pd
def filter_rows(df):
    filtered_df = df[(df['A'] > 5) & (df['B'] < 10)]
    return filtered_df
data = {
    'A': [3, 8, 6, 2, 9],
    'B': [5, 2, 9, 3, 1],
    'C': [1, 7, 4, 5, 2]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
filtered_df = filter_rows(df)
print("\nFiltered DataFrame:")
print(filtered_df)
```

```python
#Question 9
def calculate_statistics(df):
    mean_value = df['Values'].mean()
    median_value = df['Values'].median()
    std_dev = df['Values'].std()

    return mean_value, median_value, std_dev

df = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})
mean, median, std_dev = calculate_statistics(df)

print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)

#We can do di by using describe function if we have avaivle data set
```

```python
#Qustion 10
def calculate_moving_average(df):
    df = df.sort_values('Date')
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
    return df
data = {
    'Date': pd.date_range(start='2024-01-01', periods=10, freq='D'),
    'Sales': [200, 220, 250, 210, 230, 240, 210, 260, 280, 300]
}
df = pd.DataFrame(data)
calculate_moving_average(df)
```

```python
#question 11
def add_weekday_column(df):
    df['Date'] = pd.to_datetime(df['Date'])
    df['Weekday'] = df['Date'].dt.day_name()
    return df
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']
}
df = pd.DataFrame(data)
modified_df = add_weekday_column(df)
modified_df
```

```python
#Question 12
def select_date_range(df, start_date, end_date):
    df['Date'] = pd.to_datetime(df['Date'])
    mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
    return df.loc[mask]
data = {
    'Date': pd.date_range(start='2022-12-25', periods=15, freq='D')
}
df = pd.DataFrame(data)
filtered = select_date_range(df, '2023-01-01', '2023-01-31')
filtered
```
